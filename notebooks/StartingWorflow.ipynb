## Weather Products

* GSOD Daily Interpolation Weather Product  
    - 'daily-weather:gsod-interpolated:v0'  
    - 10km aprox
    - bands = ['tavg', 'tmax', 'tmin', 'rh', 'prec']
    - Description: interpolated raster from 1980-01-01 for geographical area from -180 deg to 180 deg longitude, and from -60 to 60 deg latitude.
    - URL: https://catalog.descarteslabs.com/?/product/daily-weather:gsod-interpolated:v0

* GOES-17 (16)
    - 'goes17:fulldisk:v1'
    - different res: 500 meters --  2000 meters (band depedent)

    - bands = ['derived:evi', 'derived:ndvi', 'derived:ndwi[,1,2]']
    - other bands: swir1: snow/ice
    - Comparison GOES-X: https://cimss.ssec.wisc.edu/satellite-blog/archives/29001

    - 'chirps:daily:v1'
